```python
import plotly.graph_objects as go
import plotly.io as pio
from datetime import datetime
import requests
import time

# Helper function to remove null values from data and timestamps
def remove_null_values(data, timestamps):
    clean_data = []
    clean_timestamps = []
    for i in range(len(data)):
        if data[i] is not None:
            clean_data.append(data[i])
            clean_timestamps.append(timestamps[i])
    return clean_data, clean_timestamps

# Function to fetch stock data from Yahoo API
def fetch_stock_data(stock_code):
    yahoo_url = f'https://tw.stock.yahoo.com/_td-stock/api/resource/FinanceChartService.ApacLibraCharts;autoRefresh=1715224690029;symbols=%5B"{stock_code}.TW"%5D;type=tick?bkt=&device=desktop&ecma=modern&feature=enableGAMAds%2CenableGAMEdgeToEdge%2CenableEvPlayer&intl=tw&lang=zh-Hant-TW&partner=none&region=TW&site=finance&tz=Asia%2FTaipei&ver=1.4.102&returnMeta=true'

    try:
        response = requests.get(yahoo_url)
        if response.status_code == 200:
            data = response.json()['data'][0]['chart']['indicators']['quote'][0]['close']
            timestamps = response.json()['data'][0]['chart']['timestamp']
            return remove_null_values(data, timestamps)
        else:
            print(f"Error: {response.status_code}")
            return None, None
    except Exception as e:
        print(f"Error fetching stock data: {e}")
        return None, None

# Function to plot stock data using Plotly with improved styling
def plot_stock_data(stock_code, data, timestamps):
    # Convert Unix timestamps to readable dates
    dates = [datetime.utcfromtimestamp(ts) for ts in timestamps]

    fig = go.Figure()

    # 添加價格走勢線，調整線條顏色和寬度
    fig.add_trace(go.Scatter(x=dates, y=data, mode='lines', name=stock_code,
                             line=dict(color='royalblue', width=3)))

    # 調整圖表的背景、網格、標題、座標軸樣式
    fig.update_layout(
        title={
            'text': f'{stock_code} 即時行情',
            'y':0.9,
            'x':0.5,
            'xanchor': 'center',
            'yanchor': 'top',
            'font': {
                'family': "Arial",
                'size': 24,
                'color': "#333333"
            }
        },
        xaxis_title='時間',
        yaxis_title='價格',
        xaxis=dict(
            gridcolor='rgba(200, 200, 200, 0.4)',
            tickangle=-45,
            tickformat='%Y-%m-%d %H:%M',
            showline=True,
            linewidth=2,
            linecolor='black',
            title_font=dict(size=18),
        ),
        yaxis=dict(
            gridcolor='rgba(200, 200, 200, 0.4)',
            showline=True,
            linewidth=2,
            linecolor='black',
            title_font=dict(size=18),
        ),
        font=dict(family="Arial", size=14, color="#333333"),
        paper_bgcolor='white',
        plot_bgcolor='rgba(245, 245, 245, 1)',  # 調整背景顏色
        margin=dict(l=50, r=50, t=80, b=50),
    )

    # 添加網格和圖表框線
    fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='lightgrey')
    fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='lightgrey')

    fig.show()

# Main function to update stock data periodically if during trading hours
def update_stock_data(stock_code):
    current_date = datetime.now()
    current_day = current_date.weekday()  # 0 = Monday, 6 = Sunday
    current_hour = current_date.hour
    current_minute = current_date.minute

    # Check if it's during trading hours (Mon-Fri, 9:00-13:30)
    if current_day >= 0 and current_day <= 4 and current_hour >= 9 and (current_hour < 13 or (current_hour == 13 and current_minute <= 30)):
        print("Fetching stock data during trading hours...")
        while current_day >= 0 and current_day <= 4 and current_hour >= 9 and (current_hour < 13 or (current_hour == 13 and current_minute <= 30)):
            data, timestamps = fetch_stock_data(stock_code)
            if data and timestamps:
                plot_stock_data(stock_code, data, timestamps)
            time.sleep(3600)
    else:
        print("Fetching stock data outside trading hours...")
        data, timestamps = fetch_stock_data(stock_code)
        if data and timestamps:
            plot_stock_data(stock_code, data, timestamps)

# 輸入股票代號進行測試
stock_code = input()  # 替換為你想要查詢的股票代號
update_stock_data(stock_code)
```

```python
import requests
import json

# Yahoo Finance 的股票 API，替換 {stock_code} 為實際股票代號
def fetch_stock_data(stock_code):
    # 替換 {stock_code} 為你要查詢的股票代號，例如 0050
    target_url = f'https://tw.stock.yahoo.com/_td-stock/api/resource/FinanceChartService.ApacLibraCharts;autoRefresh=1715224690029;symbols=%5B"{stock_code}.TW"%5D;type=tick?bkt=&device=desktop&ecma=modern&feature=enableGAMAds%2CenableGAMEdgeToEdge%2CenableEvPlayer&intl=tw&lang=zh-Hant-TW&partner=none&prid=1lvf1qtj3og0r&region=TW&site=finance&tz=Asia%2FTaipei&ver=1.4.102&returnMeta=true'

    try:
        # 發送 GET 請求
        response = requests.get(target_url)

        # 確認請求是否成功
        if response.status_code == 200:
            data = response.json()  # 轉換為 JSON 格式
            # 格式化輸出結果
            print(json.dumps(data, indent=4, ensure_ascii=False))
        else:
            print(f"請求失敗，狀態碼: {response.status_code}")
    except Exception as e:
        print(f"請求發生錯誤: {e}")

# 輸入股票代號
stock_code = input("請輸入股票代號: ")

# 抓取股票數據
fetch_stock_data(stock_code)
```

```python
# 確保你已安裝 plotly
!pip install plotly
!pip install -U kaleido

import requests
import plotly.graph_objects as go
import plotly.io as pio
import os
from io import BytesIO
from PIL import Image
import IPython.display as display

# 假設 Yahoo API 返回的股票數據
def get_stock_data(stock_code):
    # 模擬 Yahoo API 的數據
    if stock_code.upper() == "0050":
        return {
            "timestamps": ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05'],
            "prices": [140, 142, 141, 145, 143]
        }
    elif stock_code.upper() == "AAPL":
        return {
            "timestamps": ['2024-10-01', '2024-10-02', '2024-10-03', '2024-10-04', '2024-10-05'],
            "prices": [175, 178, 180, 179, 182]
        }
    else:
        return None

# 繪製股票走勢圖
def generate_stock_chart(stock_code):
    stock_data = get_stock_data(stock_code)

    if stock_data:
        # 繪製走勢圖
        fig = go.Figure()
        fig.add_trace(go.Scatter(x=stock_data['timestamps'], y=stock_data['prices'], mode='lines', name=stock_code))
        fig.update_layout(title=f"{stock_code} 股價走勢", xaxis_title="日期", yaxis_title="價格")

        # 將圖表轉換為圖片 (PNG 格式)
        img_bytes = pio.to_image(fig, format='png')

        # 定義保存路徑
        file_path = f'{stock_code}_stock_chart.png'

        # 將圖片保存到本地
        with open(file_path, 'wb') as f:
            f.write(img_bytes)

        print(f"股票走勢圖已保存為 {file_path}")
        return file_path

    print("無法找到該股票代碼，請確認輸入正確")
    return None

# 讓使用者輸入股票代碼
stock_code = input("請輸入股票代號: ")

# 生成並顯示股票走勢圖
chart_path = generate_stock_chart(stock_code)

# 在 Colab 中顯示圖片
if chart_path:
    image = Image.open(chart_path)
    display.display(image)
```
